```python
# Importing the required packages for data manipulation and visualization
import numpy as np        # Importing NumPy for numerical operations
import pandas as pd       # Importing Pandas for data manipulation and analysis
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting
import seaborn as sns     # Importing Seaborn for statistical data visualization

# Setting Seaborn's default style for the plots
sns.set()

# Importing the PercentFormatter from Matplotlib's ticker module to convert values into percentage format
from matplotlib.ticker import PercentFormatter
```

### Bar Chart - Creating Bar Chart of Ice Cream Shop Sales

```python
# Reading the ice cream shop sales data into pandas dataframe
df_bar_chart = pd.read_csv('bar_chart_data.csv',
                            delimiter = ',',
                            skipfooter = 1) 
```

```python
# Displaying the bar chart data
df_bar_chart
```

```python
# Changing the background of the bar graph to white using seaborn
sns.set_style('white')

# Setting the size of the plot using matplotlib
plt.figure(figsize=(8, 6))

# Plotting the bar chart defining the axes and bar width
plt.bar(df_bar_chart['Cities'],        # Cities on the x-axis
        df_bar_chart['Frequency'],     # Frequency on the y-axis
        width=0.4)                     # Setting the bar width

# Labelling the x-axis with a font size of 14 and bold font weight
plt.xlabel('Cities', fontsize=14, fontweight='bold')

# Labelling the y-axis with a font size of 14 and bold font weight
plt.ylabel('Frequency', fontsize=14, fontweight='bold')

# Increasing the font size of the x-axis labels to 13
plt.xticks(fontsize=13)

# Providing a title to the plot with a font size of 18 and bold font weight
plt.title('Sales', fontsize=18, fontweight='bold')

# Removing the top and right borders of the chart
sns.despine()

# Saving the bar graph as a PNG file
plt.savefig('Sales.png')

# Displaying the bar chart
plt.show()
```

### Area Chart - Creating an Area Chart of a Company's Revenue

```python
# Reading the company revenue data into pandas dataframe
df_area_chart_data = pd.read_csv('area_chart_data.csv')
```

```python
# Displaying the data
df_area_chart_data 
```

```python
# Setting the size of the plot using matplotlib
plt.figure(figsize=(12, 8))

# Saving a seaborn color palette Set1
pal = sns.color_palette("Set1")

# Saving the labels for the legend
labels = ['International', 'Domestic', 'Services']

# Plotting the area chart providing the date on the x-axis and the revenue columns on the y-axis
plt.stackplot(df_area_chart_data['Date'],      # Date on the x-axis
              df_area_chart_data['International'],  # International revenue
              df_area_chart_data['Domestic'],       # Domestic revenue
              df_area_chart_data['Services'],       # Services revenue
              colors=pal,                           # Using the saved color palette
              edgecolor='none')                     # No edge color for the areas

# Providing the legend with the labels and setting its location to the upper left
plt.legend(labels=labels, loc='upper left')

# Labelling the x-axis and increasing the font size to 13
plt.xlabel('Date', fontsize=13)

# Labelling the y-axis and increasing the font size to 13
plt.ylabel('Revenue (in thousand $)', fontsize=13)

# Rotating the x-axis labels to 45 degrees
plt.xticks(rotation=45)

# Providing the title, increasing the font size, and making it bold
plt.title('Revenue in 2020', fontsize=16, fontweight='bold')

# Removing the top and right borders of the chart
sns.despine()

# Saving the full chart as a PNG file, ensuring the entire chart is included
plt.savefig('Revenue in 2020.png', bbox_inches='tight')

# Displaying the plot
plt.show()
```

### Histogram - Creating a Histogram Showing Distribution of Age

```python
# Reading the histogram survey data as a pandas dataframe
df_histogram_data = pd.read_csv('histogram_survey_data.csv')
```

```python
# Displaying the histogram survey data
df_histogram_data
```

```python
# Checking the minimum and maximum age in the dataframe
min_age = df_histogram_data['Age'].min()
max_age = df_histogram_data['Age'].max()
```

```python
# Setting the plot background style to white using seaborn
sns.set_style('white')

# Setting the plot size using matplotlib
plt.figure(figsize=(8, 6))

# Plotting the histogram with age on the x-axis, 12 bins, and setting color to seagreen
plt.hist(df_histogram_data['Age'],
         bins=12,
         color='seagreen')

# Using numpy to set the x-axis ticks, increasing by 5 from the minimum to the maximum age
plt.xticks(np.arange(min(df_histogram_data["Age"]),
                     max(df_histogram_data["Age"]),
                     5.0))

# Labelling the x-axis with a font size of 12
plt.xlabel('Age Distribution', fontsize=12)

# Labelling the y-axis with a font size of 12
plt.ylabel('Number of Participants', fontsize=12)

# Providing a title to the plot with a font size of 16 and bold font weight
plt.title('Participants by Age', fontsize=16, fontweight='bold')

# Removing the top and right borders from the plot
sns.despine()

# Saving the histogram as a PNG file
plt.savefig('Participants by Age.png')

# Displaying the plot
plt.show()
```

### Regression plot - Regression plot of Auto Insurance in Sweden

```python
# Reading the regression plot data into pandas dataframe
df_auto_insurance = pd.read_csv('regression_auto_insurance_sweden.csv')
```

```python
# Displaying the regression plot data
df_auto_insurance
```

```python
# Setting the figure size using seaborn
sns.set(rc={'figure.figsize': (9, 6)})

# Setting the plot background style to white
sns.set_style('white')

# Plotting the regression plot using seaborn
sns.regplot(data=df_auto_insurance,  # Specifying the dataframe to use
            x=df_auto_insurance['Claims'],  # Specifying the x-axis data
            y=df_auto_insurance['Payment'], # Specifying the y-axis data
            scatter_kws={'color': 'black'}, # Setting the color of the scatter points to black
            line_kws={'color': 'red'})      # Setting the color of the regression line to red

# Providing a title to the plot with a larger font size and bold font weight
plt.title('Car Insurance Claims and Payment in Sweden',
          fontsize=16,
          fontweight='bold')

# Saving the regression plot as a PNG file
plt.savefig('Car Insurance Claims and Payment in Sweden.png')

# Displaying the plot
plt.show()
```

### Bar and line chart - Pareto Chart of Movie Goers' Complaints

```python
# Reading the movie goers' data into pandas dataframe
df_movie_goers_complaints = pd.read_csv('bar_line_chart.csv')
```

```python
# Displaying the movie goers' data
df_movie_goers_complaints
```

```python
# Creating a figure and a set of subplots with a specific figure size
fig, ax = plt.subplots(figsize=(11, 8))

# Creating a bar plot on the primary y-axis (ax) for the number of complaints
ax.bar(df_movie_goers_complaints['Movie Theater Goers Complaints'],
       df_movie_goers_complaints['Number of complaints'],
       color='g')  # Setting the bar color to green

# Setting the y-axis label for the primary y-axis with a larger font size
ax.set_ylabel('Number of Complaints', fontsize='large')

# Customising the y-axis ticks for the primary y-axis
ax.tick_params(axis='y',
               width=2,          # Setting the width of the ticks
               labelsize='large') # Setting the font size of the tick labels

# Creating a secondary y-axis sharing the same x-axis
ax1 = ax.twinx()

# Setting the y-axis limits for the secondary y-axis
ax1.set_ylim(0, 1.1)

# Formatting the secondary y-axis to display percentages
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1.0))

# Plotting the frequency line on the secondary y-axis
ax1.plot(df_movie_goers_complaints['Movie Theater Goers Complaints'],
         df_movie_goers_complaints['frequency'],
         color='blue',  # Setting the line color to blue
         marker='D')    # Setting the marker style to diamond

# Setting the y-axis label for the secondary y-axis with a larger font size and blue color
ax1.set_ylabel('Frequency', color='blue', fontsize='large')

# Customising the y-axis ticks for the secondary y-axis
ax1.tick_params(axis='y',
                width=2,             # Setting the width of the ticks
                labelsize='large',   # Setting the font size of the tick labels
                labelcolor='blue',   # Setting the color of the tick labels
                color='blue')        # Setting the color of the ticks

# Setting the x-axis label with a larger font size
ax.set_xlabel('Type of Complaints', fontsize='large')

# Setting the title of the plot with a larger font size and bold font weight
ax.set_title('Movie Goers Complaint types with number of complaints and the frequency',
             fontsize='x-large',
             fontweight='bold')

# Saving the plot as a PNG file
plt.savefig('Movie Goers Complaint types with number of complaints and the frequency.png')

# Displaying the plot
plt.show()
```
